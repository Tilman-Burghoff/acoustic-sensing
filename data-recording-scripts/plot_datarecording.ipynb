```python
def small_box():
    xs = -1.5704 + 2.4 + np.arange(8) * 0.1
    ys = -0.3 - 1.4 - np.arange(6) * 0.1
    dir = 1
    for y in ys:
        for x in xs[::dir]:
            yield x,y
        dir *= -1
```
