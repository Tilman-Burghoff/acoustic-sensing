```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("results_KNN.csv", index_col=0)
```

```python
# std of error in deg for better readibility
df.loc[df.Metricname == "Standard Deviation of Error", "Result"] = np.rad2deg(df.loc[df.Metricname == "Standard Deviation of Error"].Result)
```

```python
df.loc[df.Result.isna()]
```

```python
tabledata = pd.pivot_table(df, index='Modelname', columns='Metricname', values='Result') # aggregate means
```

```python
tabledata.style.apply(lambda col: ['font-weight:bold' if x==col.min() else '' for x in col])
```

```python
ax = sns.barplot(data=df.loc[df.Metricname == "Mean Square Error"], x="Modelname", y="Result", errorbar=("ci", 95))
plt.xticks(rotation=45)
plt.ylabel("Mean Square Error [Rad]")
plt.xlabel("")
plt.title("Mean Square Error")
```

```python
# df.loc[df.Metricname == "Standard Deviation of Error", "Result"] = np.rad2deg(df.loc[df.Metricname == "Standard Deviation of Error"].Result)

ax = sns.barplot(data=df.loc[df.Metricname == "Standard Deviation of Error"], x="Modelname", y="Result", errorbar=("ci", 95))
plt.xticks(rotation=45)
plt.ylabel("Standard Deviation [deg]")
plt.xlabel("")
plt.title("Standard Deviation of Error")
```

```python
ax = sns.barplot(data=df.loc[df.Metricname == "Standard deviation per Recording (avg)"], x="Modelname", y="Result", errorbar=("ci", 95))
plt.xticks(rotation=45)
plt.ylabel("Standard Deviation [rad]")
plt.xlabel("")
plt.title("Standard Deviation per Recording averaged")
```

```python
ax = sns.barplot(data=df.loc[df.Metricname == "R squared"], x="Modelname", y="Result", errorbar=("ci", 95))
plt.xticks(rotation=45)
plt.ylabel("R squared")
plt.xlabel("")
plt.title("R squared")
```

```python
ax = sns.barplot(data=df.loc[df.Metricname == "Ratio of Outliers (>= 0.5 rad)"], x="Modelname", y="Result", errorbar=("ci", 95))
plt.xticks(rotation=45)
plt.ylabel("Ratio of Outliers (Error >= 0.5 rad)")
plt.xlabel("")
plt.title("Ratio of Outliers (Error >= 0.5 rad)")
```
