```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_outside = pd.read_csv("results_mic_outside.csv")
df_contact0 = pd.read_csv("results_contact_joint0.csv")
df_contact3 = pd.read_csv("results_contact_joint3.csv")
```

```python
df_outside[["Position"]] = "Outside"
df_contact0[["Position"]] = "Joint 0"
df_contact3[["Position"]] = "Joint 3"
```

```python
df = pd.concat([df_outside, df_contact0, df_contact3])
# std of error in deg for better readibility
df.loc[df.Metricname == "Standard Deviation of Error", "Result"] = np.rad2deg(df.loc[df.Metricname == "Standard Deviation of Error"].Result)
```

```python
ax = sns.barplot(data=df.loc[df.Metricname == "Mean Square Error"], x="Modelname", y="Result", errorbar=("ci", 95), hue="Position")
plt.xticks(rotation=45)
plt.ylabel("Mean Square Error [Rad]")
plt.xlabel("")
plt.title("Mean Square Error")
```

```python
ax = sns.barplot(data=df.loc[(df.Metricname == "Mean Square Error") & (df.Joint == 0) & ~(df.Modelname == "kNN")], x="Modelname", y="Result", errorbar=("ci", 95), hue="Position")
plt.xticks(rotation=45)
plt.ylabel("Mean Square Error [Rad]")
plt.xlabel("")
plt.title("Mean Square Error at joint 0")
```

```python
ax = sns.barplot(data=df.loc[(df.Metricname == "Mean Square Error") & (df.Joint == 3) & ~(df.Modelname == "kNN")], x="Modelname", y="Result", errorbar=("ci", 95), hue="Position")
plt.xticks(rotation=45)
plt.ylabel("Mean Square Error [Rad]")
plt.xlabel("")
plt.title("Mean Square Error at joint 3")
```

```python
ax = sns.barplot(data=df.loc[(df.Metricname == "Mean Square Error") & ~(df.Modelname == "kNN")], x="Joint", y="Result", errorbar=("ci", 95), hue="Position")
plt.ylabel("Mean Square Error [Rad]")
plt.xlabel("")
plt.title("Mean Square Error averaged over Models (w/o kNN)")
```

```python
ax = sns.barplot(data=df.loc[(df.Metricname == "Standard Deviation of Error") & ~(df.Modelname == "kNN")], x="Modelname", y="Result", errorbar=("ci", 95), hue="Position")
plt.xticks(rotation=45)
plt.ylabel("Standard Deviation [deg]")
plt.xlabel("")
plt.title("Standard Deviation of Error")
```

```python
ax = sns.barplot(data=df.loc[(df.Metricname == "Standard Deviation of Error") & (df.Joint ==0) & ~(df.Modelname == "kNN")], x="Modelname", y="Result", errorbar=("ci", 95), hue="Position")
plt.xticks(rotation=45)
plt.ylabel("Standard Deviation [deg]")
plt.xlabel("")
plt.title("Standard Deviation of Error at Joint 0")
```
