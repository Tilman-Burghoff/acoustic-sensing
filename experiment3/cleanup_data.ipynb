```python
import pandas as pd
```

```python
data = pd.read_csv("data/samples.csv")
```

```python
q_0_to_small = data.loc[data.q_0 < -1.6].idx
q_0_to_small
```

```python
recorded_twice = data.loc[(data.q_0 > 0.8) & (data.q_0 < 1.3) & (data.q_3 < -2.29) & (data.q_3 > -2.31) & (data.session_id > 4), "idx"]
recorded_twice
```

```python
illegal_indizes = pd.concat([q_0_to_small, recorded_twice])
```

```python
list(illegal_indizes.index)
```

```python
df = data.loc[~data.idx.isin(list(illegal_indizes.index))]
```
