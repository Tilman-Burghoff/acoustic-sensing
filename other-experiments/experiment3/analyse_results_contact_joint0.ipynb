```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("results_contact_joint0.csv")
```

```python
# std of error in deg for better readibility
df.loc[df.Metricname == "Standard Deviation of Error", "Result"] = np.rad2deg(df.loc[df.Metricname == "Standard Deviation of Error"].Result)
```

```python
df.loc[df.Result.isna()]
```

```python
np.unique(df.Modelname)
```

```python
order = ['kNN', 'Linear', '3-Layer FC', '4-Layer FC', '5-Layer FC', '1-Channel CNN', '2-Channel CNN',  '3-Channel CNN','4-Channel CNN']
tabledata = pd.pivot_table(df, index='Modelname', columns=['Metricname', "Joint"], values='Result') # aggregate means
tabledata = tabledata.reindex(order)
```

```python
tabledata.style.apply(lambda col: ['font-weight:bold' if x==col.min() else '' for x in col])
```

```python
ax = sns.barplot(data=df.loc[df.Metricname == "Mean Square Error"], x="Modelname", y="Result", errorbar=("ci", 95), order=order)
plt.xticks(rotation=75)
plt.ylabel("Mean Square Error [Rad]")
plt.xlabel("")
plt.title("Mean Square Error")
```

```python
ax = sns.barplot(data=df.loc[df.Metricname == "Mean Square Error"], x="Modelname", y="Result", errorbar=("ci", 95), hue="Joint")
plt.xticks(rotation=45)
plt.ylabel("Mean Square Error [Rad]")
plt.xlabel("")
plt.title("Mean Square Error")
```
